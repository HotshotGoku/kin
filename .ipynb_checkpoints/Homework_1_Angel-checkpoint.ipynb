### b. Calculate the covariance matrix of the data and interpret the result. 

```python
#Normalizing data
data_norm = pd.DataFrame(stats.zscore(data, axis = 1), index = data.index, columns = data.columns)
data_norm
#covariance with only normalized data (mxm, genes x genes)
covariance_matrix = np.cov(data_norm,bias=True)
print(covariance_matrix)
#changing values of diagonals to zero
np.fill_diagonal(covariance_matrix, 0)
print(covariance_matrix)

#converting nump py covariance_matrix to dataframe as cov_df
cov_df = pd.DataFrame(covariance_matrix, index = data_norm.index, columns = data_norm.index)
cov_df
abs_df = cov_df.abs()
stacked_matrix=abs_df.stack().sort_values(ascending = False)
#top 10 varying gene pairs
stacked_matrix.head(10)
#correlating biological functions to top 10 genes. 
genes = ['b3517', 'b1493','b1493','b3517', 'b0552', 'b3505','b3505',  'b0552', 'b0072',  'b0073','b0073' , 'b0072', 'b3993',  'b3991','b3991' , 'b3993', 'b3505' , 'b2192', 'b2192',  'b3505']
samples = ['operon', 'cog']
gene_info.loc[genes, samples]
```

### d. Calculate principal components and component weights by calculating the eigendecomposition of the covariance matrix of your data. Compare to the result of a PCA package to ensure your result. Be sure to use standardized data (zero mean, unit variance for each variable across samples).

```python
#Finding eigen values and eigen vectors
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
Real_eig_vecs=eig_vecs.real
Real_eig_vecs

#Projecting the data on eigen vectors
XV_matrix=np.matmul(data_norm.T, Real_eig_vecs)

#Plotting the PC1 and PC2
plt.figure(figsize=(8,6))
plt.scatter(XV_matrix.values[:,0],XV_matrix.values[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title("PCA through eigendecomposition")

#Comparing with PCA package in sklearn
pca=PCA()
pca.fit(data_norm.T)
x_pca=pca.transform(data_norm.T)
#Plotting PC1 vs PC2
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title("PCA comparison")
```

## 3. Comparing Factorization Methods

### Read “Enter the Matrix: Factorization Uncovers Knowledge from Omics” by Stein-O’Brien et. al. https://www.cell.com/trends/genetics/fulltext/S0168-9525(18)30124-0. Contrast PCA, NMF, and ICA in their use in analyzing biological data. 

Briefly, PCA finds dominant sources of variation in high-dimensional datasets, inferring genes that distinguish between samples. PCA finds factors in rows of the pattern matrix that can be ranked by the amount of variation that they explain in the data, as illustrated in a scree plot. PCA analyses typically plot the first two principal components (PCs; rows of the pattern matrix) to assess sample clustering. Maximizing the variability captured in specific factors, as opposed to spreading relatively evenly among factors, may mix the signal from multiple CBPs in a single component. Therefore, PCA may conflate processes that sometimes occur and complicate interpretation of the amplitude matrix for defining data-driven gene sets or the inference of specific CBPs. 

ICA and NMF learn distinct processes from an input data matrix using different techniques. ICA learns factors that are statistically independent, resulting in more accurate association with literature-derived gene sets. ICA finds factors associated with independent sources of variation, and therefore cannot be ranked in a scree plot. The relative absolute value of the magnitude of each element in the pattern matrix indicates the extent to which that sample contributes to the corresponding source of variation. 

NMF methods constrain all elements of the amplitude and pattern matrices to be greater than or equal to zero. Whereas the features in PCA can be ranked by the extent to which they explain the variation in the data, the features in both ICA and NMF are assumed to have equal weight. NMF is well suited to transcriptional data, which is typically non-negative itself, and semi-NMF is also applicable to data that can have negative values. The assumptions of NMF model both the additive nature of CBPs and parsimony, generating solutions that are biologically intuitive to interpret. NMF finds factors that are both non-negative and not ranked by relative importance, similarly to ICA. The value of the pattern matrix indicates the extent to which each sample contributes to an inferred source of variation and is associated with overexpression of corresponding gene weights in the amplitude matrix. 

In contrast to PCA, the components of ICA and NMF cannot be ranked by percent variation explained. Instead, each row of the pattern matrix is an equally important CBP in the data. Therefore, these patterns are plotted independently relative to biological covariates, and not relative to one another as in PCA analyses.
