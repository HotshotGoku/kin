#### A.A. K-means

1. **Overall criteria:** For a given k, k-means attempts to divide the data points into k subsets, such that the sum of cluster variances is minimized (i.e. sum of the squared distances to the centroid for each cluster is minimized).

    **Algorithm:** 
    1. Select K points as initial centroids, usually by selecting k random points
    2. Form K clusters by assigning each point to its closest centroid
    3. Compute the centroids of each cluster based on its points
    4. Repeat steps B and C until the centroids do not change
    5. Usually repeat steps A-D using different initial points, then select the clustering that separates the data best

2. **Parameters:**
    - K, the number of clusters

3. **Strengths and weaknesses:**
    * Strengths: Simple, computationally inexpensive, requires only one parameter
    * Weaknesses: Difficult to determine K for many datasets, doesn't work well on non-spherical or loose clusters (which may be split into artificial subclusters), susceptible to local minima, susceptible to outliers dramatically shifting centroids
    


#### A.B. Agglomerative hierarchical clustering

1. **Overall criteria:** This algorithm creates a hierarchy of clusters by iteratively merging the two closest clusters until all points have been merged. It can therefore generate a dendrogram and mark subclusters down to single data points.

    **Algorithm:** 
    1. Initialize all points to their own cluster
    2. Generate a proximity matrix between all clusters using a chosen distance metric
        1. *Single linkage:* minimal distance between points in the clusters - creates chains of nearest neighbors and loose clusters
        2. *Complete linkage:* maximal distance between points in the clusters - creates compact clusters, but allows for some points to be closer to other clusters than their own (crowding).
        3. *Average linkage:* average distance - removes crowding and chaining problems, but makes the proximity measure at the end difficult to interpret
    3. Merge the two closest clusters
    4. Continue until all points are in the same cluster
2. **Parameters:**
     - Chosen linkage/distance metric
     - A way to create clusters: either a distance threshold or cluster number

3. **Strengths and weaknesses:** 
    * Strengths: Can be deterministic (as long as ties are broken consistently), handles variable cluster size well, robust to outliers, can incorporate categorical variables through custom distance metrics, provides a useful dendrogram
    * Weaknesses: Computationally expensive, difficult to tell where to cut the dendrogram into discrete clusters, results vary significantly by linkage metric chosen

#### A.C. DBSCAN

1. **Overall criteria:** Points are clustered with their nearest neighbors such that the density of each cluster is relatively constant. Sparse regions become noise/background and are not assigned to clusters.

    **Algorithm:** 
    1. Compute the density of each point, which is the number of points within a radius *eps* of the point (self-inclusive).
    2. Classify each point as:
        * Core: interior of a dense region, density >= *minPts*
        * Border: not a core point, but within *eps* of a core point
        * Noise/background: not a core or border point
    3. Eliminate all noise
    4. Connect all core points that are within *eps* of each other
    5. Assign each border point to its nearest cluster
2. **Parameters:**
    - *eps*, the radius that defines a point's neighborhood
    - *minPts*, the threshold point number to define a dense region in a neighborhood

3. **Strengths and weaknesses:**
    * Strengths: Great at separating clusters with odd shapes, no need to select the number of clusters beforehand, excellent outlier handling, can be deterministic, Fairly computationally inexpensive
    * Weaknesses: Struggles with uneven cluster densities, requires tuning of *eps* and *minPts*.

#### A.D. Gaussian mixture models

1. **Overall criteria:** This algorithm views data as a mixture of K gaussian distributions (multivariate normals) and determines the probabilities of each point being in each distribution (cluster)
   
   **Algorithm:** 
    1. Choose the # of distributions to fit. Initialize guesses for these distributions and their weights (i.e. with K-means).
    2. E step: Compute the expected likelihood of observing the data, weighted with current parameter estimates (often called "responsibilities" of each point
    3. M step: Update parameter estimates to maximize the expected likelihood from the E step. Iterate E/M steps until expected likelihood converges
    4. Assign clusters based on the distribution most likely to have generated for each point
    
2. **Parameters:**
    - K, the number of distributions to use
    
3. **Strengths and weaknesses:** 
    * Strengths: Works well with variable cluster sizes and densities, provides a probability instead of categorical assignment (useful for outlier identification)
    * Weaknesses: Requires a K parameter and will use all K regardless of appropriate cluster number, sensitive to initialization, Doesn't work well with elliptical, non-linear, or other non-gaussian data
    

### B. Calculate the k-means clusters for Homework_Data for different numbers of clusters. How similar are the clustering patterns? How do the clusters align with sample conditions?

### <span style="color:red">20 points</span>

#### Calculate the clusters

```python
# we will cluster the normalized data so that genes get an equal say in cluster assignment
data_norm = pd.DataFrame(stats.zscore(data, axis=1), index = data.index, columns = data.columns)
```

```python
# I love dataframes - store all cluster labels in one df,[samples x k]
kmeans_labels = pd.DataFrame(columns = [2, 5, 11, 55], index = data.columns)
# must choose at least three values of k, anything in the interval [2, 55]

# compute the clusters
for k in kmeans_labels.columns:
    kmeans = KMeans(n_clusters = k, random_state = 1).fit(data_norm.T)
    
    # store the labels in a column of kmeans_labels
    kmeans_labels[k] = kmeans.labels_
kmeans_labels
```

#### How similar are the clustering patterns?

```python
# use PCA to visualize the clusters
pca = PCA()
data_pc = pca.fit_transform(data_norm.T)
data_pc = pd.DataFrame(data_pc, index = data.columns)

# plot everything
fig, ax = plt.subplots(2, 2, figsize = (12, 9))

# I want the subplots to form a square. This will help me index into the 2D array of axes
ax_ind = [[0, 0], [0, 1], [1, 0], [1, 1]]

for a, k in zip(ax_ind, kmeans_labels.columns): # gets an axis index for plotting, and the k value
    
    current_axis = ax[a[0]][a[1]]
    
    # plot each cluster separately. There are k clusters
    for c in range(k):
        samples = kmeans_labels.index[kmeans_labels[k] == c]
        current_axis.scatter(data_pc.loc[samples, 0], data_pc.loc[samples, 1])
    
    # add labels
    current_axis.set_title('K = %i'%(k))
```

For K = 2, 5, and 11, the clusters are fairly distinct on the PCA plot. They are visually pretty similar.

We observe the blue cluster on the left of K = 1 break into 4 parts for K = 5. It is interesting to track the other cluster, the samples which are labeled in orange in K = 2:
* K = 5: two samples have moved out of this cluster, which must indicate similarities which are not visible in the first two PCs.
* K = 11: the original cluster is now three clusters, and one of them (green) also includes many samples that were labeled blue in K = 2.
Since the clusters are not simply subdividing but are exchanging samples, the borders between clusters at k = 2 are fuzzy and unreliable. This is to be expected for K-means, since it is deterministic and based on centroids instead of intrinsic relationships in the data.

```python
# align clusters with sample conditions, option 2: Visualization

# make a dataframe that will become our heatmaps
# In this case, use the fraction of study samples in each cluster (The recall)
# Note that other measures would also be accepted.

metadata_col = 'Study'

# for incomplete metadata, this replaces empty space with a value (-1)
# prevents future errors
metadata = metadata.fillna(-1)

clust_study_df = pd.DataFrame(columns = metadata[metadata_col].unique())

# loop through all clusters. These are rows of the heatmap.
for k in kmeans_labels.columns: # all k
    for clust in range(k): # all clusters
        
        # name of this cluster
        row_name = 'k = %i; Cluster %i'%(k, clust) 
        # get the members of this cluster
        clust_samples = kmeans_labels.index[kmeans_labels[k] == clust]
        # use sets because they make math easy
        clust_samples = set(clust_samples)
        
        # loop through all studies. These are columns of the heatmap.
        for study in clust_study_df.columns:
            
            # get the study samples as a set
            study_samples = metadata.index[metadata[metadata_col] == study]
            study_samples = set(study_samples)
            
            # compute the fraction: [# samples in both]/[# samples in study]
            both = clust_samples.intersection(study_samples)
            fraction = len(both)/len(study_samples)
            
            # save in the dataframe
            clust_study_df.loc[row_name, study] = fraction

# we will get a type error if we don't fix this:
clust_study_df = clust_study_df.astype(float)
            
# draw the plots
fig, ax = plt.subplots(4, figsize = (10, 12))

for a, k in zip(ax, kmeans_labels.columns): # gets an axis for plotting and the k value
    
    # find the rows of interest (pythonically - post to piazza if you want to learn this syntax)
    rows = [r for r in clust_study_df.index if 'k = %i;'%(k) in r]
    
    # show the heatmap
    a.imshow(clust_study_df.loc[rows], aspect = 'auto', cmap = 'Blues')
    
    # make it pretty
    a.set_ylabel('K = %i'%(k))
    if k != 55: # 55 is a lot of ticks...
        a.set_yticks(range(k))
        a.set_xticks([]) # turn off labels for the heatmaps that aren't at the bottom
    else: # add labels for the bottom heatmap
        a.set_xticks(range(clust_study_df.shape[1]))
        a.set_xticklabels(clust_study_df.columns)
```

This visualization is a little better at showing how the clusters match the metadata since it provides a holistic view. There are better, but less easy to explain measures of cluster overlap than this fractional approach as well. 

Down each column, we either see either (1) a single dark blue element which indicates a perfect overlap between study and cluster or (2) several lighter blue elements, where the study was split between them.

We can see which studies are made up of a single compact cluster this way: Base and Acid are always together, even at high K. OmpR is also fairly distinct -- we can see it even comprises its own cluster when K = 11 (cluster 0).

Overall, study does not seem to be the defining characteristic for these clusters. By changing the variable 'metadata_col', you can try different features and see if any of them align. For example, 'pH' will show two columns: one for the acid samples and one without. In my analysis, k-means can't be fully explained by any of the metadata but shows some alignment.

### A. Method descriptions

### <span style="color:red">15 points</span>

There are many approaches to evaluate how well your data clusters (e.g. a simple method is to order samples by clusters and simply visualize the similarity matrix). Discuss how the following methods can be used to validate/evaluate cluster (Note: They do not all serve the same purpose):

A) Silhouette coefficient

B) Resampling of data (bootstrapping/cross-validation)

C) Ensemble modeling




#### A.A. Silhouette coefficient
* $s_i = (b_i - a_i)/max(a_i, b_i)$, where $s_i$ is the silhouette coefficient of sample $i$, $a_i$ is the average distance from $i$ to all other samples in its cluster and $b_i$ is the minimum (over all clusters) average distance to points in another cluster.
* Larger SC indicates that the intercluster distance > cluster diameter
* Perfect separation: SC = 1; Perfect overlap: SC = -1.
* You can examine SC distributions within clusters, take the average SC of the clusters, or the average SC of all of the data.

#### A.B. Resampling of the data (bootstrapping/cross-validation)
* By using subsets of the data (for example, clustering the samples while only looking at a random fraction of the genes), we can identify more robust clusters.
* Cross-validation: split the data into k subsets and cluster k times leaving out one subset each time. This is a more rigorous way to obtain robust clusterings.
* Bootstrapping: sample the original dataset randomly with replacement (essentially shuffling the numbers of the matrix around) to create a random dataset and see how well that clusters. This can demonstrate the underlying null distribution of the dataset and be used to compute confidence intervals.

#### A.C. Ensemble modeling
* Generate clusters using various methods (see 1A - also vary parameters if desired) and find a consensus set of clusters
* This is very robust because it will easily detect erroneous clusterings
* However, some data will naturally be clustered best by a given algorithm (see the strengths in 1A)

### C. Calculate and interpret the silhouette coefficients after clustering your data using k-means clustering with k = 5.

### <span style="color:red">15 points</span>

```python
# compute the silhouette coefficients
labels = kmeans_labels[5]
SC = metrics.silhouette_samples(data_norm.transpose(), labels)

# generate a histogram
fig, ax = plt.subplots()
ax.hist(SC, bins = 30, range = [-1, 1])
ax.set_xlabel('Count')
ax.set_ylabel('Silhouette Coefficient')
```

From this histogram, we can see that the SCs are normally distributed slightly above 0. In general, that means the clusters are nearly borderline between 'good', well-separated clusters, and 'bad' -- but they lean toward good. We could try ensemble clustering to look for ways of improving this.

```python
# Look at SCs on a per-cluster basis

# seaborn wants the data in this shape for the swarm plot
swarm_df = pd.Series(SC, index = data.columns).to_frame(name = 'SC')
swarm_df['cluster'] = labels

fig, ax = plt.subplots(1, 2, figsize = (10,5))

# make the swarm plot
sns.swarmplot(x='cluster', y='SC', data = swarm_df, ax = ax[0])

# put it next to a biplot
for c in range(5):
    samples = kmeans_labels.index[kmeans_labels[5] == c]
    ax[1].scatter(data_pc.loc[samples, 0], data_pc.loc[samples, 1], label = 'cluster %i'%(c))
ax[1].legend()
```

From the swarm plot, we can see that our 'best', or most unique cluster, is the purple cluster 4: it has the highest SC and its samples are visibly far from the others on the PCA plot. We can also see the set of the worst samples, the 4 elements in cluster 1 that have negative SCs and look like they are right next to the red and blue clusters (near (-20, 20)) on the biplot.
